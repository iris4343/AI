```python
def Heavenly_Pool():
    Tianchi_data = {
        '定位': '在胸部，當第4肋間隙，乳頭外1寸，前正中線旁開5寸',
        '解剖': '在胸大肌外下部，胸小肌下部起端，深層為第四肋間內、外肌；有胸腹壁靜脈，胸外側動、靜脈分支；布有胸前神經肌支及第四肋間神經。',
        '主治': '胸悶，心煩，咳嗽，痰多，氣喘，胸痛，腋下腫痛，瘰癧，瘧疾，乳癰。',
        '配伍': '配列缺、豐隆治咳嗽；配內關治心痛；配支溝治脅肋痛。',
        '刺灸法': '斜刺或平刺0.5～0.8寸；可灸。本穴正當胸腔，內容心、肺，不宜深刺。'
    }
    return Tianchi_data

def Marsh_at_the_Crook():
    Quze_data = {
        '定位': '在肘橫紋中，當肱二頭肌腱的尺側緣',
        '解剖': '在肱二頭肌腱的尺側；當肱動、靜脈處；布有正中神經的本干',
        '主治': '心痛，善驚，心悸，胃疼，嘔吐，轉筋，熱病，煩躁，肘臂痛，上肢顫動，咳嗽。',
        '配伍': '配神門、魚際治嘔血；配內關、大陵治心胸痛；配大陵、心俞、厥陰俞治心悸、心痛；配少商、尺澤、曲池治療肘臂攣急、肩臂痛。',
        '刺灸法': '直刺0.8～1寸，或者用三棱針刺血；可灸。'
}
    return Quze_data

def Heavenly_Spring():
    Tianquan_data = {
        '定位': '在臂內側，當腋前紋頭下2寸，肱二頭肌的長、短頭之間。',
        '解剖': '在肱二頭肌的長短頭之間；有肱動、靜脈肌支；為臂內側皮神經及肌皮神經分布處。',
        '主治': '心痛，胸脅脹滿，咳嗽，胸背及上臂內側痛。',
        '配伍': '配內關、通里治心痛、心悸；配肺俞、支溝治咳嗽、胸脅痛；配俠白、曲池、外關治上肢痿、痹、癱、痛。',
        '刺灸法': '直刺0.5～0.8寸；可灸。'
    }
    return Tianquan_data

def Inner_Pass():
    Neiguan_data = {
        '定位': '在前臂掌側，當曲澤與大陵的連線上，腕橫紋上2寸，掌長肌腱與橈側腕屈肌腱之間',
        '解剖': ' 在橈側腕屈肌腱與掌長肌腱之間，有指淺屈肌，深層為指深屈肌；有前臂正中動、靜脈，深層為前臂掌側骨間動、靜脈；布有前臂內側皮神經，下為正中神經掌皮支，最深層為前臂掌側骨間神經。',
        '主治': '心痛，心悸，胸痛，胃痛，嘔吐，呃逆，失眠，癲狂，癇證，郁證，眩暈，中風，偏癱，哮喘，偏頭痛，熱病，產後血暈，肘臂攣痛。',
        '配伍': '配公孫治肚痛；配膈俞治胸滿支腫；配中脘、足三里治胃脘痛、嘔吐、呃逆；配外關、曲池治上肢不遂、手振顫。配患側懸厘治偏頭痛；配建里除胸悶',
        '刺灸法': ' 直刺0.5～1寸；可灸。'
    }
    return Neiguan_data

def Xi_Cleft_Gate():
    Ximen_data = {
        '定位': '在前臂掌側，當曲澤與大陵的連線上，腕橫紋上5寸。',
        '解剖': '在橈側腕屈肌腱與掌長肌腱之間，有指淺屈肌，深部為指深屈肌；有前臂正中動、靜脈，深部為前臂掌側骨間動、靜脈；布有前臂內側皮神經，其下為正中神經，深層有前臂掌側骨間神經。',
        '主治': ' 心痛，心悸，胸痛，心煩，咳血，嘔血，衄血，疔瘡，癲疾。',
        '配伍': '配大陵止咯血；配曲澤、大陵治心痛；配梁丘、足三里、太沖治神經性嘔吐；配內關治急性缺血性心肌損傷。',
        '刺灸法': '直刺0.5～1寸；可灸'
    }
    return Ximen_data

def Pericardium_Meridian_of_Hand_Jueyin():
    acupoints_info = {
        '天池穴': Heavenly_Pool(),
        '曲澤穴': Marsh_at_the_Crook(),
        '天泉穴': Heavenly_Spring(),
        '內關穴': Inner_Pass(),
        '郄門穴': Xi_Cleft_Gate()
        # ... 其他穴位
    }
    return acupoints_info

def get_acupoint_info(acupoint_name, info_category=None):
    acupoints_info = Pericardium_Meridian_of_Hand_Jueyin()

    acupoint_data = acupoints_info.get(acupoint_name, None)
    if acupoint_data is None:
        return "未找到指定的穴位信息。"

    if info_category:
        return acupoint_data.get(info_category, "未找到指定類別的信息。")

    return acupoint_data

def query_acupoint():
    while True:
        print("\n--- 查詢系統 ---")
        acupoint_name = input("請輸入您要查詢的穴位名稱（輸入'退出'結束查詢）:\n(天池穴、雲門穴、天泉穴、內關穴、郄門穴選擇其一):")
        if acupoint_name == "退出":
            print("\n--- 已退出查詢 ---")
            print("感謝使用本系統，祝你好運owo")
            break

        acupoint_info = get_acupoint_info(acupoint_name)
        if "未找到" in acupoint_info:
            print(f"\n{acupoint_info}\n")
            continue

        while True:
            print(f"\n找到穴位：{acupoint_name}")
            print("可查詢的信息類別有：定位、解剖、主治、配伍、刺灸法")
            info_category = input("請輸入您要查詢的信息類別（輸入'全部'顯示所有信息，或輸入'返回'返回上一層）:")

            if info_category == "返回":
                print("\n返回上一層...\n")
                break
            elif info_category == "全部":
                print(f"\n{acupoint_name} - 全部信息:")
                for category, detail in acupoint_info.items():
                    print(f"{category}: {detail}")
                print()  # 為了排版整潔owo
            elif info_category in ["定位", "解剖", "主治", "配伍", "刺灸法"]:
                category_info = get_acupoint_info(acupoint_name, info_category)
                print(f"\n{acupoint_name} - {info_category}:\n{category_info}\n")
            else:
                print("\n輸入的信息類別不正確，請重新輸入。\n")

            # 再次查詢或返回
            while True:
                next_action = input("輸入'再查詢'以查詢此穴位的其他信息，或輸入'返回'返回穴位選擇：")
                if next_action in ["再查詢", "返回"]:
                    break
                else:
                    print("請輸入有效的指令（'再查詢' 或 '返回'）。")
            if next_action == "返回":
                print("\n返回穴位選擇...")
                break
# 啟動查詢
query_acupoint()
```
